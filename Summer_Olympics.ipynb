# Summer Olympics Data Analysis Assignment

### 1. In how many cities Summer Olympics is held so far?

```python
import pandas as pd
```

```python
df = pd.read_csv("summer.csv")
```

```python
len(set(df['City']))
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
data=[]

df=df[df['Medal']=='Gold']

for sport in df['Sport'].unique():
    data.append([sport,len(df[df['Sport']==sport])])

pd.DataFrame(data,columns=['Sport','Medals']).sort_values(by='Medals',ascending=False).head().plot(x='Sport',y='Medals',kind='bar',figsize=(6,4))
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
data = []

for Sport in df['Sport'].unique():
    data.append([Sport , len(df[df['Sport']  == Sport])])

pd.DataFrame(data,columns = ['Sport','freq']).sort_values(by='freq', ascending=False).head().plot(x = 'Sport', y = 'freq', kind = 'bar', figsize = (6,4))
```

### 4. Which player has won most number of medals? (Top 5)

```python
data = []

for Athlete in df['Athlete'].unique():
    data.append([Athlete , len(df[df['Athlete']  == Athlete])])

pd.DataFrame(data,columns = ['Athlete','freq']).sort_values(by='freq', ascending=False).head().plot(x = 'Athlete', y = 'freq', kind = 'bar', figsize = (6,4))
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
data=[]

df=df[df['Medal']=='Gold']

for Athlete in df['Athlete'].unique():
    data.append([Athlete,len(df[df['Athlete'] == Athlete])])

pd.DataFrame(data,columns=['Athlete','Medals']).sort_values(by='Medals',ascending=False).head().plot(x='Athlete',y='Medals',kind='bar',figsize=(6,4))
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
data = []

df=df[df['Country']=='IND']
df=df[df['Medal']=='Gold']
df=df.sort_values(by='Year',ascending=True).head(1)
print("In the year ",df['Year'])
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
data = []

for Event in df['Event'].unique():
    data.append([Event , len(df[df['Event']  == Event])])

pd.DataFrame(data,columns = ['Event','freq']).sort_values(by='freq', ascending=False).head().plot(x = 'Event', y = 'freq', kind = 'bar', figsize = (6,4))
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
pl_gold = df[df["Medal"]=="Gold"]
pl_female_gold = pl_gold[df['Gender']=='Women']
pl_female_gold = pl_female_gold.groupby("Sport").count()["Medal"].sort_values(ascending=False).head()
pl_female_gold.plot.bar(figsize=(10,5))
```
